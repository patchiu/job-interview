```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('ssa_fb.csv')  
df
```

```python
df=df.replace('?', np.nan).replace('$', np.nan).replace(')', np.nan).fillna(method='ffill',axis=1)
df
```

```python
df= df.loc[:,['Unnamed: 2','Unnamed: 6','Unnamed: 9']]
df=df.drop([n for n in range(4)])
```

```python
df.columns = ['', 'June 30, 2018','December 31, 2017']
df=df.dropna().reset_index(drop=True)
df
```

```python
df.iloc[0:2,1:3]=np.nan
df.iloc[12:14,1:3]=np.nan
df.iloc[21:23,1:3]=np.nan
df.iloc[24,1]='(687)'
df.iloc[24,2]='(227)'
df
```

```python

df.iloc[24,1]=687
df.iloc[24,2]=227

df['June 30, 2018'] = df['June 30, 2018'].str.replace(',', '')
df['December 31, 2017'] = df['December 31, 2017'].str.replace(',', '')

df['June 30, 2018'] = df['June 30, 2018'].apply(lambda x: float(x) if x != np.nan else(x))
df['December 31, 2017'] = df['December 31, 2017'].apply(lambda x: float(x) if x != np.nan else(x))

df
```
